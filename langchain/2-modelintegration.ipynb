### Streaming And Batch

#### Streaming

Most models can stream their output content while it is being generated. By displaying output progressively, streaming significantly improves user experience, particularly for longer responses. Calling `stream()` returns an iterator that yields output chunks as they are produced. You can use a loop to process each chunk in real-time:


```python
model.invoke("write me a 200 words essay on the importance of AI in modern society.")
```

```python
for chunk in model.stream("write me a 200 words essay on the importance of AI."):
    print(chunk.text, end="|", flush=True)
```

### Batch

Batching a collection of independent requests to a model can significantly improve performance and reduce costs, as the processing can be done in parallel.

```python
responses = model.batch(
    [
        "Why do parrots have the ability to mimic human speech?",
        "How do mobile phones work?",
        "How do airplanes fly?",
        "Why do animals have different colors?",
    ],
    config={
        # max_concurrency will limit the number of concurrent requests to 2 means out of these 4 questions at a time only 2 questions
        # will be processed and the other 2 will wait for their turn to be processed
        "max_concurrency": 2,
    }
)

for r in responses:
    print(r.content)
```
